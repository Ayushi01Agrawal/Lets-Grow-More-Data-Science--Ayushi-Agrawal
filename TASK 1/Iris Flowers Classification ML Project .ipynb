# Building Model , Training and Testing

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['Species_Flower'] = le.fit_transform = (df['Species_Flower'])
df.head(5)
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(columns=['Species_Flower'])
Y = df['Species_Flower']
x_train , x_test , y_train , y_test = train_test_split(X , Y , test_size = 0.3)
```

# 1. Logistic Regression

```python
# Initialize a Logistic Regression
lg= LogisticRegression(max_iter=1000)
```

```python
lg.fit(x_train,y_train)
```

```python
# Predict on the test set and calculate accuracy
y_pred=lg.predict(x_test)
score=accuracy_score(y_test,y_pred)
```

```python
def report(model):
    preds=model.predict(x_test)
    print(classification_report(preds,y_test))
    plot_confusion_matrix(model,x_test,y_test)
```

```python
print('Logistic Regression')
report(lg)
print(f'Accuracy: {round(score*100,2)}%')
```

# 2. Linear SVC

```python
rbf_sv= SVC()
rbf_sv.fit(x_train,y_train)
L_svc=LinearSVC()
```

```python
L_svc.fit(x_train,y_train)
```

```python
# Predict on the test set and calculate accuracy
y_pred=L_svc.predict(x_test)
score=accuracy_score(y_test,y_pred)
```

```python
print('Linear SVC')
report(L_svc)
print(f'Accuracy: {round(score*100,2)}%')
```

# 3. Decision Tree Classifier


```python
DTC = DecisionTreeClassifier()
```

```python
DTC=DTC.fit(x_train,y_train)
```

```python
y_pred=DTC.predict(x_test)
score=accuracy_score(y_test,y_pred)
```

```python
print('Decision Tree Classifier')
report(DTC)
print(f'Accuracy: {round(score*100,2)}%')
```

# 4. Multinomial Naive Bayes

```python
NB= MultinomialNB()
```

```python
NB.fit(x_train,y_train)
```

```python
y_pred=NB.predict(x_test)
score=accuracy_score(y_test,y_pred)
```

```python
print('NB')
report(NB)
print(f'Accuracy: {round(score*100,2)}%')
```

# 5. KNN

```python
KNN=KNeighborsClassifier(n_neighbors=6)
```

```python
KNN.fit(x_train, y_train)
```

```python
y_pred=KNN.predict(x_test)
score=accuracy_score(y_test,y_pred)
```

```python
print('KNN')
report(KNN)
print(f'Accuracy: {round(score*100,2)}%')
```
